### Valores Nulos

```python
# Contar valores nulos por columna
missing_values = df.isnull().sum().sort_values(ascending=False)
missing_percentage = (missing_values / len(df)) * 100

# Crear DataFrame con resumen
missing_df = pd.DataFrame({
    'Valores Nulos': missing_values,
    'Porcentaje (%)': missing_percentage
})

# Mostrar las columnas con nulos
print("🔍 Columnas con valores nulos:\n")
print(missing_df[missing_df['Valores Nulos'] > 0])

# Visualización
plt.figure(figsize=(10, 6))
missing_df[missing_df['Valores Nulos'] > 0]['Porcentaje (%)'].plot(kind='barh', color='coral')
plt.title('Porcentaje de valores nulos por columna')
plt.xlabel('Porcentaje (%)')
plt.ylabel('Columnas')
plt.tight_layout()
plt.show()
```

#### Análisis de valores nulos

Observaciones clave:

- Varias columnas tienen **valores faltantes significativos**, especialmente:
  - `Sunshine` (~48% nulos)
  - `Evaporation` (~43% nulos)
  - `Cloud9am`, `Cloud3pm` (~38–40% nulos)
  - `Pressure9am` y `Pressure3pm` (~10% nulos)
- Variables como `MinTemp`, `MaxTemp`, `Humidity`, `WindSpeed` y `Rainfall` también tienen nulos, pero en menor proporción (1–2%).

Conclusión:

- Las columnas con más del 40% de datos faltantes podrían ser excluidas o imputadas con mucho cuidado.
- Para variables numéricas clave y moderadamente afectadas, se propone **imputar con la mediana**, ya que es robusta ante outliers y distribuciones sesgadas.


```python
# Imputar columnas numéricas con mediana
num_cols = df.select_dtypes(include='number').columns
for col in num_cols:
    if df[col].isnull().sum() > 0:
        mediana = df[col].median()
        df[col].fillna(mediana, inplace=True)

# Imputar columnas categóricas con moda
cat_cols = df.select_dtypes(exclude='number').columns
for col in cat_cols:
    if df[col].isnull().sum() > 0:
        moda = df[col].mode()[0]
        df[col].fillna(moda, inplace=True)

# Verificar que no queden nulos
print(df.isnull().sum())
```

#### Imputación de valores faltantes

Tratamiento de valores nulos con mediana

**Acción realizada:**

Se imputaron los valores faltantes de todas las **columnas numéricas** utilizando la **mediana** de cada columna.

**¿Por qué la mediana?**

- Es una **medida robusta** frente a outliers, a diferencia de la media que se ve afectada por valores extremos.
- Muchas variables como `Rainfall` y `RISK_MM` tienen distribuciones **sesgadas**, por lo que la mediana representa mejor el valor central típico.
- Ayuda a mantener la distribución original de los datos sin generar sesgos importantes.

**Resultado:**

- Se eliminaron los nulos de las columnas numéricas sin eliminar filas completas.
- El dataset ahora es más adecuado para el modelado posterior (como regresión o clasificación).

### Outliers

```python
# Solo columnas numéricas
numeric_cols = df.select_dtypes(include='number').columns

# Boxplots para detectar outliers
for col in numeric_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col], color='lightblue')
    plt.title(f'Boxplot de {col}')
    plt.xlabel(col)
    plt.tight_layout()
    plt.show()
```

#### Análisis de Outliers

**Observaciones desde los boxplots:**

- Variables como `Rainfall`, `Evaporation`, `Sunshine`, `WindGustSpeed`, y `RISK_MM` muestran una gran cantidad de **valores extremos altos**.
- Esto es común en fenómenos meteorológicos: la mayoría de los días tienen valores bajos, pero en ciertos eventos (lluvias intensas, ráfagas de viento) se presentan valores atípicos.
- Las temperaturas (`MinTemp`, `MaxTemp`, `Temp9am`, `Temp3pm`) presentan algunos outliers, pero menos pronunciados.

**Conclusión:**

- Los **outliers extremos** pueden distorsionar modelos sensibles (como regresión lineal), por lo que es recomendable **tratarlos adecuadamente**, especialmente en variables que serán utilizadas para predicción.
- En este caso, se aplicará el **método del rango intercuartílico (IQR)** para detectar y recortar valores extremos.


```python
# Aplicar tratamiento de outliers por IQR (recorte)
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    # Recortar valores fuera del rango
    df[col] = df[col].clip(lower, upper)

# Verificar resultados con resumen estadístico
print("📊 Estadísticas después del tratamiento de outliers:\n")
print(df[numeric_cols].describe())
```

#### Tratamiento de Outliers con IQR

**Acción realizada:**

Se aplicó el método de **rango intercuartílico (IQR)** para detectar outliers y se utilizó la función `clip()` para **recortar** (no eliminar) los valores extremos fuera del rango [Q1 - 1.5×IQR, Q3 + 1.5×IQR].

**¿Por qué este enfoque?**

- **No elimina filas**, por lo que no se pierde información general del dataset.
- Evita que valores extremos distorsionen estadísticas como la media o afecten modelos sensibles.
- Es **más conservador** que eliminar registros completos, y mantiene la estructura del dataset.

**Resultado:**

- Las variables numéricas ahora están dentro de un rango razonable sin valores atípicos extremos.
- Esto facilita un mejor desempeño en modelos predictivos y análisis posteriores.

### Tratamiento de categóricos

```python
# Transformar columnas binarias: RainToday y RainTomorrow
# (Yes = 1, No = 0)
binarias = ['RainToday', 'RainTomorrow']
df[binarias] = df[binarias].apply(lambda x: x.map({'No': 0, 'Yes': 1}))

# Codificar variables no binarias con One-Hot Encoding
# (Location, WindGustDir, WindDir9am, WindDir3pm)
df = pd.get_dummies(df, columns=['Location', 'WindGustDir', 'WindDir9am', 'WindDir3pm'], drop_first=True)

# Mostrar forma final del dataset
print("✅ Dataset codificado. Nueva forma:", df.shape)
```

#### Codificación de variables categóricas
**¿Qué se hizo?**

1. Se codificaron las variables binarias `RainToday` y `RainTomorrow` como:
   - `Yes` → 1
   - `No` → 0
   Esto es útil para modelos de clasificación binaria.
2. Se aplicó **One-Hot Encoding** a variables con múltiples categorías:
   - `Location` (49 ciudades)
   - Direcciones de viento (`WindGustDir`, `WindDir9am`, `WindDir3pm`)
   Se usó `drop_first=True` para evitar la **multicolinealidad**.
**¿Por qué esta estrategia?**

- **Label Encoding** sólo se usó en variables binarias, donde el orden no afecta.
- Para variables nominales con múltiples categorías, **One-Hot Encoding** es más apropiado porque no impone un orden arbitrario.
- La transformación convierte todo el dataset a formato **totalmente numérico**, esencial para algoritmos como regresión, árboles, redes neuronales, etc.

**Resultado:**
