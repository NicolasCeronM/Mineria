#### Anlisis de valores nulos

Observaciones clave:

- Varias columnas tienen **valores faltantes significativos**, especialmente:
  - `Sunshine` (~48% nulos)
  - `Evaporation` (~43% nulos)
  - `Cloud9am`, `Cloud3pm` (~3840% nulos)
  - `Pressure9am` y `Pressure3pm` (~10% nulos)
- Variables como `MinTemp`, `MaxTemp`, `Humidity`, `WindSpeed` y `Rainfall` tambin tienen nulos, pero en menor proporcin (12%).

Conclusin:

- Las columnas con ms del 40% de datos faltantes podran ser excluidas o imputadas con mucho cuidado.
- Para variables numricas clave y moderadamente afectadas, se propone **imputar con la mediana**, ya que es robusta ante outliers y distribuciones sesgadas.


```python
# Imputar columnas numricas con mediana
num_cols = df.select_dtypes(include='number').columns
for col in num_cols:
    if df[col].isnull().sum() > 0:
        mediana = df[col].median()
        df[col].fillna(mediana, inplace=True)

# Imputar columnas categricas con moda
cat_cols = df.select_dtypes(exclude='number').columns
for col in cat_cols:
    if df[col].isnull().sum() > 0:
        moda = df[col].mode()[0]
        df[col].fillna(moda, inplace=True)

# Verificar que no queden nulos
print(df.isnull().sum())
```

#### Imputacin de valores faltantes

Tratamiento de valores nulos con mediana

**Accin realizada:**

Se imputaron los valores faltantes de todas las **columnas numricas** utilizando la **mediana** de cada columna.

**Por qu la mediana?**

- Es una **medida robusta** frente a outliers, a diferencia de la media que se ve afectada por valores extremos.
- Muchas variables como `Rainfall` y `RISK_MM` tienen distribuciones **sesgadas**, por lo que la mediana representa mejor el valor central tpico.
- Ayuda a mantener la distribucin original de los datos sin generar sesgos importantes.

**Resultado:**

- Se eliminaron los nulos de las columnas numricas sin eliminar filas completas.
- El dataset ahora es ms adecuado para el modelado posterior (como regresin o clasificacin).

### Outliers

```python
# Solo columnas numricas
numeric_cols = df.select_dtypes(include='number').columns

# Boxplots para detectar outliers
for col in numeric_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col], color='lightblue')
    plt.title(f'Boxplot de {col}')
    plt.xlabel(col)
    plt.tight_layout()
    plt.show()
```

#### Anlisis de Outliers

**Observaciones desde los boxplots:**

- Variables como `Rainfall`, `Evaporation`, `Sunshine`, `WindGustSpeed`, y `RISK_MM` muestran una gran cantidad de **valores extremos altos**.
- Esto es comn en fenmenos meteorolgicos: la mayora de los das tienen valores bajos, pero en ciertos eventos (lluvias intensas, rfagas de viento) se presentan valores atpicos.
- Las temperaturas (`MinTemp`, `MaxTemp`, `Temp9am`, `Temp3pm`) presentan algunos outliers, pero menos pronunciados.

**Conclusin:**

- Los **outliers extremos** pueden distorsionar modelos sensibles (como regresin lineal), por lo que es recomendable **tratarlos adecuadamente**, especialmente en variables que sern utilizadas para prediccin.
- En este caso, se aplicar el **mtodo del rango intercuartlico (IQR)** para detectar y recortar valores extremos.


```python
# Aplicar tratamiento de outliers por IQR (recorte)
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    # Recortar valores fuera del rango
    df[col] = df[col].clip(lower, upper)

# Verificar resultados con resumen estadstico
print(" Estadsticas despus del tratamiento de outliers:\n")
print(df[numeric_cols].describe())
```

#### Tratamiento de Outliers con IQR

**Accin realizada:**

Se aplic el mtodo de **rango intercuartlico (IQR)** para detectar outliers y se utiliz la funcin `clip()` para **recortar** (no eliminar) los valores extremos fuera del rango [Q1 - 1.5IQR, Q3 + 1.5IQR].

**Por qu este enfoque?**

- **No elimina filas**, por lo que no se pierde informacin general del dataset.
- Evita que valores extremos distorsionen estadsticas como la media o afecten modelos sensibles.
- Es **ms conservador** que eliminar registros completos, y mantiene la estructura del dataset.

**Resultado:**

- Las variables numricas ahora estn dentro de un rango razonable sin valores atpicos extremos.
- Esto facilita un mejor desempeo en modelos predictivos y anlisis posteriores.
